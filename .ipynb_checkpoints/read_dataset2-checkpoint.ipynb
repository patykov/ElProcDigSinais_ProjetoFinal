```python
import os
import time
import numpy as np
from sklearn.covariance import GraphLassoCV, ledoit_wolf
import matplotlib.pyplot as plt
```

```python
### READING DATASET ###

# Paths
dataset_path = 'Dataset'
games_dir = next(os.walk(dataset_path))[1]

tag_to_skip = [
    'highlights', 'highlights2', '.DS_Store', '.git', 
    'em_mcs_energy_diff_ascending', 'em_mcs_energy_diff', 'em_mcs_energy']

# Iterating through games
games_data = []
games_path = [os.path.join(dataset_path, g) for g in games_dir]
for path in games_path:
    game_name = os.path.basename(path)
    if game_name != 'ned_svk':
        game_info = os.listdir(path)
        game_data = dict()
        for info in game_info:
            tag_name = os.path.splitext(info)[0]
            if tag_name not in tag_to_skip:
                csv_file = os.path.join(path, info)
                if tag_name == 'highlights_boundaries':
                    data_value = np.genfromtxt(csv_file, delimiter=',', dtype=None)
                else:
                    data_value = np.fromfile(csv_file, dtype=np.float64)
                game_data[tag_name] = data_value
        game_data['name'] = game_name
        games_data.append(game_data)

games_backup = games_data[:]
```

```python
### CRIANDO VETOR DE GT ###

wanted_cases = ['Gol', 'Perigo']
to_skip = ['name', 'highlights_boundaries']

for game in games_data:
    h = game['highlights_boundaries']
    end = h[-1][0]
    gt = np.zeros(end)
    for line in h:
        tag = line[2].decode('UTF-8').split(' ')
        name = [t for t in tag if t != ''][-1]
        if name in wanted_cases:
            gt[line[0]-1:line[1]] = 1                   
            
    game['gt'] = gt
    del game['highlights_boundaries']
```

```python
### ANTES DE CORTAR OS VETORES ###

for game in games_data:
    print(game['name'])
    values = [len(v) for k, v in game.items() if k not in ['name']]
    min_l = min(values)
    max_l = max(values)
    print('\tMin: {}  Max: {}  Diff: {}'.format(min_l, max_l, max_l-min_l))
```

```python
### CORTANDO VALORES EXCEDENTES PARA QUE TODOS OS DADOS TENHAM O MESMO TAMANHO ###

for game in games_data:
    values = [len(v) for k, v in game.items() if k not in ['name']]
    min_l = min(values)
    for k, v in game.items():
        if len(v) > min_l:
            game[k] = v[:min_l]
            

### ZERO PADDING ###
max_length = 0
for game in games_data:
    values = [len(v) for k, v in game.items() if k not in ['name']]
    max_length = max(values) if max(values)>max_length else max_length

for game in games_data:
    for k, v in game.items():
        if k not in ['name']:
            zero_v = np.zeros(max_length)
            if len(v) < max_length:
                zero_v[:len(v)] = v
                game[k] = zero_v
```

```python
### CHECANDO COMPRIMENTO DOS DADOS ###

for game in games_data:
    print(game['name'])
    values = [len(v) for k, v in game.items() if k not in ['name', 'highlights_boundaries']]
    min_l = min(values)
    max_l = max(values)
    print('\tMin: {}  Max: {}  Diff: {}'.format(min_l, max_l, max_l-min_l))
```

```python
for g_id, g in enumerate(games_data):
    print('\n\n{}-{}'.format(g_id, g['name']))
    X = []
    X_labels = []
    for k, v in g.items():
        if k not in ['name', 'highlights_boundaries']:
            X.append(v)
            X_labels.append(k)

# for i, tag_name in enumerate(games_data[10].keys()):
#     if tag_name == 'name':
#         name_id = i
        
# X = []
# for game in games_data:
#     values_array = np.array(list(game.values()))

#     # Removing name and GT to train
#     values_array = np.delete(values_array, (name_id), axis=0)

#     values_array = np.array(values_array[:].tolist())
#     if len(X) == 0:
#         X = np.transpose(values_array)
#     else:
#         X = np.concatenate((X, np.transpose(values_array)))

    X = np.array(X)
    # number of rows in one chunk
    SPLITROWS = 1000

    numrows = X.shape[0]

    # subtract means form the input data
    X -= np.mean(X, axis=1)[:,None]

    # normalize the data
    X /= np.sqrt(np.sum(X*X, axis=1))[:,None]

    # reserve the resulting table onto HDD
    res = np.memmap("/tmp/mydata.dat", 'float64', mode='w+', shape=(numrows, numrows))

    for r in range(0, numrows, SPLITROWS):
        for c in range(0, numrows, SPLITROWS):
            r1 = r + SPLITROWS
            c1 = c + SPLITROWS
            chunk1 = X[r:r1]
            chunk2 = X[c:c1]
            res[r:r1, c:c1] = np.dot(chunk1, chunk2.T)
       
    
    ### Identificando dados correlacionados ###
    for l, line in enumerate(res):
        line = np.round_(line[l:], 4)
        indices = [i+l for i,v in enumerate(line) if v ==1.0]
        if len(indices) > 1:
            print('\t({})-{:22} e ({}){:22} tem correlacao 1'.format(
                    indices[0], X_labels[indices[0]], indices[1], X_labels[indices[1]]))

        indices = [i for i,v in enumerate(line) if v ==-1.0]
        if len(indices) > 1:
            print('({})-{:22} e ({})-{:22} tem correlacao -1'.format(
                    indices[0], X_labels[indices[0]], indices[1], X_labels[indices[1]]))
```

```python
# Plot the results
plt.figure(figsize=(10, 6))

# plot the covariances
vmax = res.max()
plt.imshow(res, interpolation='nearest', vmin=-vmax, vmax=vmax,
           cmap=plt.cm.RdBu_r)
ticks = np.arange(0,numrows,1)
plt.xticks(ticks, X_labels, rotation=90)
plt.yticks(ticks, X_labels)

plt.title('Covariance matrix')
plt.show()
```
